- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
import pandas as pd
import scipy.stats as st

# Load the data
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")

# Filter HP for Dragon and Grass types (using Type 1)
dragon_hp = df[df['Type 1'] == 'Dragon']['HP']
grass_hp = df[df['Type 1'] == 'Grass']['HP']

# Calculate basic statistics
print(f"Dragon HP mean: {dragon_hp.mean():.2f}")
print(f"Grass HP mean: {grass_hp.mean():.2f}")

# Perform Independent T-Test (One-Tailed)
# 'greater' means we are testing if dragon_hp is significantly higher than grass_hp
t_stat, p_value = st.ttest_ind(dragon_hp, grass_hp, equal_var=False, alternative='greater')

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion logic
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the Null Hypothesis. Dragon Pokemons have significantly more HP than Grass Pokemons.")
else:
    print("Conclusion: Fail to reject the Null Hypothesis. There is no significant evidence that Dragon Pokemons have more HP.")
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
import pandas as pd
import scipy.stats as st

# Load the data
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")

# Define the stats to test
stats_list = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Split the data into two groups
legendary = df[df['Legendary'] == True]
non_legendary = df[df['Legendary'] == False]

print(f"{'Stat':<10} | {'T-Statistic':<12} | {'P-Value':<10} | {'Result'}")
print("-" * 55)

for stat in stats_list:
    # Perform Welch's T-test (Independent, Two-tailed)
    t_stat, p_val = st.ttest_ind(legendary[stat], non_legendary[stat], equal_var=False)
    
    # Check significance
    status = "Reject H0" if p_val < 0.05 else "Fail to reject H0"
    
    print(f"{stat:<10} | {t_stat:<12.4f} | {p_val:<10.2e} | {status}")
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
import pandas as pd
import numpy as np
import scipy.stats as st

# Load the data
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")

# 1. Function for Euclidean Distance
def get_distance(lat1, lon1, lat2, lon2):
    return np.sqrt((lat1 - lat2)**2 + (lon1 - lon2)**2)

# 2. Calculate distances to School and Hospital
school_coords = (34, -118) # Lat, Lon
hospital_coords = (37, -122) # Lat, Lon

df['dist_school'] = get_distance(df['latitude'], df['longitude'], school_coords[0], school_coords[1])
df['dist_hospital'] = get_distance(df['latitude'], df['longitude'], hospital_coords[0], hospital_coords[1])

# 3. Define closeness (Close if distance to either is < 0.5)
df['is_close'] = (df['dist_school'] < 0.5) | (df['dist_hospital'] < 0.5)

# 4. Split Dataset
close_houses = df[df['is_close'] == True]['median_house_value']
far_houses = df[df['is_close'] == False]['median_house_value']

# 5. Perform the T-test
t_stat, p_val = st.ttest_ind(close_houses, far_houses, equal_var=False, alternative='greater')

print(f"Mean Value (Close): ${close_houses.mean():,.2f}")
print(f"Mean Value (Far):   ${far_houses.mean():,.2f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Result: Reject H0 - Houses near landmarks are significantly more expensive.")
else:
    print("Result: Fail to reject H0 - No significant price difference found.")
```
